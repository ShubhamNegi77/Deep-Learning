```python
import tensorflow as tf
import numpy as np
```

```python
X = np.arange(0,10,6).astype('float')
```

```python
y = X/2+0.4
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(X, y, epochs = 100)
```

```python
model.predict([6])
```
